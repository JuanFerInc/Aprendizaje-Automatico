## 2.1 Juego


    

Teeko es un juego sencillo donde cada jugador inicia con cuatro fichas en mano y un tablero de 5x5 vacío. En la primera etapa del juego toman turnos los jugadores colocados ficha por ficha en el tablero hasta que se terminan.

Luego de que las fichas son colocadas en el tablero comienza la segunda parte del juego, esta consiste en que cada jugador toma turno para mover una de sus fichas. Cada ficha puede moverse a un casillero adyacente que esté vacío. Esto continúa hasta que uno de los jugadores gana.

Existen dos formas de ganar en el Teeko. Una de las formas es llegar a cuatro fichas en línea horizontal, vertical, o diagonal sin ninguno espacio vacío entre una ficha y otra. La segunda forma de ganar es llegar a que las cuatro fichas formen un cuadrado sin ningún espacio entre cada ficha.

Para simplificar la representación, se tomó la decisión de excluir la primera parte del juego. La partida comienza con las fichas de los jugadores posicionadas de forma aleatoria, verificando que las posiciones no son una victoria. Una partida se considera finalizada cuando o bien la distribución de las fichas da como ganador a un jugador, o bien cuando se supera cierto umbral de movimientos posibles entre ambos jugadores (500).


## 2.2 Tablero





Cada instancia del tablero puede ser representamos como una tupla de atributos con valores que permiten describir el tablero en cada jugada. Considerando que el Teeko es un juego de estrategia donde el jugador ganador es el primero que coloca las fichas en una posición ganadora, se optó por definir atributos que empujen el jugador a llevar las fichas a dicha posición, manteniéndolo informado del estado de las fichas del oponente. Se definieron atributos que describen el tablero de la siguiente forma y en siguiente orden:




- **CantFichasAdyNegro**: Calcula la cantidad de posiciones adyacentes a mis fichas que se encuentren ocupadas por mi contrincante.

- **CantFichasAdyRojo**: Análogo a calcCuantasAdyacenes, pero para el jugador rojo.

- **DispHorizontalNegro**: Se calcula la mayor distancia horizontal que hay entre dos fichas. El rango de valores va desde 0 cuando todas las fichas están en la misma columna, hasta 4 cuando hay fichas tanto en la columna 0 como en la 4.

- **DispHorizontalRojo**: Análogo a DispHorizontalNegro, pero para el jugador rojo.

- **DisVerticalNegro**: Análogo a DispHorizontalNegro pero verticalmente.

- **DisVerticalRojo**: Análogo a DisVerticalNegro pero para el jugador rojo.

- **CuadradoPotNegro**: Calcula que tan cerca están las fichas negras de formar un cuadrado. Se suma para cada ficha negra la cantidad de fichas negras adyacentes. Cuando cada ficha tiene a todas las fichas adyacente (hay un cuadrado) se obtiene el valor 12 y cuando están todas separadas se obtiene el valor 0.

- **CuadradoPotRojo**: Análogo a CuadradoPotNegro pero para el jugador rojo.

Otros atributos que fueron considerados e implementados en un principio, pero que no fueron tenidos en cuenta para la versión final:

- **MaxFichasVerticalNegro**, **MaxFichasHorizontalNegro**, **MaxFichasDiagonalNegro**: Se cuentan la cantidad de fichas que se encuentra alineada de forma vertical, horizontal y diagonal sin espacios vacíos entre si. Cuando están todas alineadas se obtiene el valor 4, cuando están todas separadas se obtiene el valor 1.

- **MaxFichasVerticalRojo**, **MaxFichasHorizontalRojo**, **MaxFichasDiagonalRojo**:Describen lo mismo que los atributos anterior pero con las fichas Rojas.

Estos son los atributos que se decidió utilizar dado que se consideró que son los que mejores describen el tablero sin tener que mencionar las posiciones de las fichas. A continuación se muestran algunos tableros y los valores de los atributos lo representan.


## 2.4 Algoritmo



A continuación describiremos las funciones utilizada en el momento de evaluar cada tablero, tomar una decisión y ajustar los pesos cuando finaliza la partida. 
La función de valoración **V(b)** calcula el valor de un tablero haciendo el producto entre el vector de pesos y el vector de atributos. Esta función obtiene el vector de atributos que describe el tablero y retorna el resultado de realizar el producto entre el vector de atributos y el vector de pesos del jugador. 


````python
def V(partida, jugador, pesos):
    atributos = jugador.calcularAtributos(partida.jugadasNegro, partida.jugadasRojo)
    total = 0

    for i, x in enumerate(atributos):
        total = total + jugador.pesos[i] * x
    telemetria.data['valoresTableros'].append(total)
    return total           
````
Inicializamos el jugador con un set de pesos que se eligen de forma aleatoria entre los valos 0.01 y 0.08. Estos pesos se utilizan para calcular el valor de cada tablero. Cuando es el turno de mover del jugador negro, calcula todas las posibles jugadas que puede realizar. Cada una de las posibles jugadas genera un nuevo tablero que es evaluado por la función de valoración. Se calculan los atributos, y se evalúan con los pesos actuales del jugador, dando un resultado entre **-1**  y **1** para cada tablero. 

La función que decide el siguiente movimiento a realizar por el jugador es la función **calularMejorProximaJugada**. Esta función calcula todos los posibles movimientos que tiene el jugador según las reglas del Teeko, evalúa cada uno de los tableros usando la función **V()**, si alguna tablero es final, se es elegido. Si no, se selecciona al azar uno de los mejores tableros que fueron evaluado. Se optó por quedarse con un umbral del 25% de los mejores tableros para realizar la selección azarosa.

La función  **actualizarPesosSegunHistorial(self, historial)** actualiza los pesos del jugador según la historia de la partida (éste contiene todos los tableros sobre los cuales el jugador a entrenar realizó su valoración y jugó más el tablero final independientemente de quién haya jugado). Los pesos se ajustan al final de cada partida. Definimos la tasa de aprendizaje de $\alpha$ = 0.001 fija dado que esta permite que los pesos se ajusten de forma gradual, evitando saltos grandes. En la sección 3.1 se muestran los experimentos que se realizaron para poder llegar a la conclusión de utilizar ese $\alpha$.

Se recorre el historial de la partida en sentido inverso (empezando por la partida final). Para el tablero final el valor de entrenamiento $Vent$ toma valores 1, -1 o 0 en caso de que haya sido una victoria, derrota o empate respectivamente. Para el resto de los tableros,  $Vent$ toma el valor de $\gamma * V(tablero siguiente)$. Se ajustan los pesos utilizando el algoritmo de mínimos cuadrados para minimizar el error cuadrático entre $Vent$ y $V$ para cada tablero del historial.


````python
    def actualizarPesosSegunHistorial(self, historial):

        errorCuadratico = 0

        viejosPesos = self.pesos
        nuevosPesos = self.pesos.copy()
        # lo doy vuelta para recorrerlo desde atras
        historial.reverse()
        for n, partida in enumerate(historial):

            # gamma = 1
            gamma = 0.9
            # gamma = 1/(n+1)

            esElTableroFinal = n == 0
            if esElTableroFinal:
                Vent = tableroFinal(partida, self)
                Vop = V(partida, self, nuevosPesos)
                errorCuadratico = (Vent-Vop)**2 + errorCuadratico
            else:
                # la diferencia ahora es que es con los atributos de siguiente tablero
                # recordar que como estamoms iterando de atras para adelante,
                # el tablero siguiente se encuentra en el indice i-1
                partidaSig = historial[n-1]
                Vent = V(partidaSig, self, nuevosPesos) * gamma
                Vop = V(partida, self, nuevosPesos)
                errorCuadratico = (Vent - Vop) ** 2 + errorCuadratico

            atributos = self.calcularAtributos(partida.jugadasNegro, partida.jugadasRojo)

            # actualizacion
            aux = nuevosPesos.copy()
            for i, w in enumerate(nuevosPesos):
                aux[i] = w + self.tasaAprendizaje * (Vent - Vop) * atributos[i]
            nuevosPesos = aux

        telemetria.data['errorCuadraticoMedio'].append(errorCuadratico)
        telemetria.data['pesos'].append(nuevosPesos.copy())

        # fin de analizar todo el historial 
        self.pesos = nuevosPesos
````


## 2.4 Contrarios


Definimos dos jugadores para nuestra evaluación de aprendizaje, para entrenar optamos por utilizar dos contrincantes diferentes. Uno de los contrincantes es un jugador que juega de forma aleatoria. Dado un tablero actual, posiciones de las fichas y posibles jugadas, el jugador selecciona una ficha al azar y un movimiento al posible al azar. Luego de tomar la decisión, realiza la jugada. A priori, se conjeturó que este jugador no sería un buen contrincante y que la experiencia adquirida del mismo no le permitiria ganarle a otros contrincantes entrenados con métodos más sofisticada.

El segundo contrincante que definimos es el jugador “sombra”. Ambos jugadores utiliza la función **calularMejorProximaJugada** para decidir qué jugada hacer. Inicialmente comienza con pesos al azar, pero luego de 50 partidas de entrenamiento, se toman lo aprendido por el jugador que está entrenando contra el jugador sombra y se pasa el conocimiento al jugador sombra; es decir, cada 50 partidas el jugador sombra recibe los pesos que tenía el otro jugador hace 50 partidas. Este proceso de “enseñar” continua a lo largo del entrenamiento, logrando así que el jugador sombra aprenda y ayude a entrenar. 

Los resultados de la partida esperados de los jugadores que entrenan son los siguientes. Que el jugador que entrena contra el jugador random tenga la mayor cantidad de victorias, aproximadamente un 80-90%. En el caso del jugador que entrena contra su sombra, se espera que tenga cerca de 50% de victorias, ya que este juega contra alguien que toma decisiones y está cerca de su nivel.

Luego de entrenar los dos jugadores, el que entrena contra el random y el que enfrenta contra la sombra, se decidió matchearlos y ver que cual de los dos resulta mejor jugador. En esta etapa ninguno de los jugadores aprenderá (sus pesos se mantendrán constantes). En esta etapa esperamos ver que el jugador que entrenó contra la sombra tenga un mayor porcentaje de victoria.

A continuacion se muestra el algoritmo que utilisamos para calcular las jugadas random.
````python
    def calcularJugadaRandom(self, partida):
        res = partida.obtenerMovimientosPosibles(self.color)
        elem = res[random.randrange(0, len(res), 1)]
        return elem['ficha'], elem['movimientosPosibles'][random.randrange(0, len(elem['movimientosPosibles']))]

````
