#### 2.1.2Preprocesamiento de los datos: valores faltantes


Luego de analizar los datos, se optó por mantener los valores fuera de dominio. tratando así a el valor ‘?’ como otro valor más del dominio de los atributos que lo contiene. 

#### 2.1.4 Preprocesamiento de los datos: atributos categóricos


Para los atributos categóricos se decidió utilizar one-hot-encoding. No se utilizó la opción de cambiar las etiquetas de un atributo por valores enteros ya que esto resultaría en un ordenamiento entre las etiquetas, inexistente en el corpus original, que no representa la realidad. Para los atributos que tenían solo dos etiquetas (como por ejemplo sex y income) se los cambió a atributos binarios.
En la columna income se cambió el valor <=50k por el 0 y el valor >50k por el 1. Esto será necesario tenerlo en cuenta al momento de interpretar los resultados obtenidos. 


```python
# One-hot-encoding para los multiclase,
# binario si no.
for cc in columnas_categoricas:
    esMulticlase = len(cc['dominio']) > 2
    columna = cc['columna']

    if esMulticlase:

        # Utilizamos scikit-learn para crear un one-hot-encoder
        ohe=sklearn.preprocessing.OneHotEncoder(sparse=False)
        # Obtenemos las categorías a partir de los datos de entrenamiento
        ohe.fit(ds[columna].to_numpy().reshape(-1,1))
        # Obtenemos los nuevos valores a partir del valor original
        new=ohe.transform(ds[columna].to_numpy().reshape(-1,1))

        for i, valor_posible in enumerate(ohe.categories_[0]):
            nomnbre_nueva_columna = columna + '_' + valor_posible
            ds[nomnbre_nueva_columna]=new[:,i]

        # la dropeamos
        ds = ds.drop([columna], axis=1)

    else:
        # si no es muliclase -> es binario
        # luego, si es binario lo codeamos con 0's y 1's

        # Creamos un labelEncoder utilizando scikit-learn
        le=sklearn.preprocessing.LabelEncoder()
        # Obtenemos las clases a partir de los valores del conjunto de entrenamiento
        le.fit(ds[columna])
        # Mostramos las clases obtenidas
        le.classes_
        # Ajustamos el campo sex, transformándolo
        ds.loc[:,columna] = le.transform(ds.loc[:,columna])

```

## 2.2 Algoritmos

### 2.2.1 Algoritmo de Naive Bayes

El algoritmo de Naïve Bayes se implementó tal cual se puede encontrar en el libro de de Machine Learning by Tom Mitchell. Para cada instancia se calcula la probabilidad de que la respuesta sea <=50k o >50k y se retorna la más verosímil.  Para evitar que se anule el término del clasificador para una instancia que no tenga ejemplos se utilizó el m-estimador. Se definió p=1/|Valores| donde |Valores| es la cantidad de posibles valores que puede tomar el atributo y con m definido luego de analizar entre varias opciones con validación cruzada.

Para evitar tener que calcular las probabilidades cada vez que se quiere evaluar una instancia, se optó por definir un diccionario de tablas donde cada atributo se guarda con su tabla de probabilidades. Estas probabilidades se calculan al inicio del algoritmo y luego se referencian cada vez que se necesita calcular la probabilidad final de cada instancia.

### 2.2.2 Algoritmo K Nearest Neighbors

El algoritmo de KNN se implementó tal cual se encuentra en las diapositivas del curso. Se permiten definir dos tipos de evaluaciones. En una se definen la cantidad de vecinos cercanos a ser considerados en el momento de evaluar una instancia. Cuanto más lejos esté el vecino de la instancia menor peso tiene su resultado. Otra opción es considerar que los k vecinos más cercano tienen el mismo peso en el momento de decidir, retornando el valor más común.

### 2.3.1 Estandarización de atributos

Como el algoritmo KNN se beneficia de que los atributos continuos tengan aproximadamente el mismo orden de magnitud, se decidió por normalizar los atributos continuos. Las medidas de media y desviación estándar se calculan sobre el conjunto de entrenamiento, y luego se utilizan para normalizar a este y a el conjunto de evaluación.

```python
media_x = X_train[columnas_numericas].mean()
std_x = X_train[columnas_numericas].std()
X_train[columnas_numericas] = (X_train[columnas_numericas] - media_x) / std_x
X_test[columnas_numericas] = (X_test[columnas_numericas] - media_x) / std_x
```

En Naïve Bayes, para calcular las probabilidades en los casos de atributos numéricos, se optó por calcular percentiles para cada una de las variables continuas y luego usar esos valores para “catalogar” a los continuos. Dando como resultado la discretización de estos bajo un nuevo dominio finito de valores enteros. Al separarlo en cuartiles se obtuvo atributos enteros en el rango [0,3]. El cálculo de los percentiles se realizó sobre el conjunto de entrenamiento, y luego se utilizaron los valores calculados para discretizar también el conjunto de entrenamiento.

```python
X_train_bayes = X_train.copy()
X_test_bayes = X_test.copy()

for col in columnas_numericas:
    percentile_25 = X_train_bayes[col].quantile(0.25)
    percentile_50 = X_train_bayes[col].quantile(0.50)
    percentile_75 = X_train_bayes[col].quantile(0.75)

    nueva_columna = 'bin_' + col # <- columnas no son binarias, toma valores [0,3]
    # agregar una sola columna con distintos valores
    X_train_bayes.loc[X_train_bayes[col] <= percentile_25, nueva_columna] = 0
    X_train_bayes.loc[(X_train_bayes[col] > percentile_25) & (X_train_bayes[col] <= percentile_50), nueva_columna] = 1
    X_train_bayes.loc[(X_train_bayes[col] > percentile_50) & (X_train_bayes[col] <= percentile_75), nueva_columna] = 2
    X_train_bayes.loc[X_train_bayes[col] > percentile_75, nueva_columna] = 3

    X_test_bayes.loc[X_test_bayes[col] <= percentile_25, nueva_columna] = 0
    X_test_bayes.loc[(X_test_bayes[col] > percentile_25) & (X_test_bayes[col] <= percentile_50), nueva_columna] = 1
    X_test_bayes.loc[(X_test_bayes[col] > percentile_50) & (X_test_bayes[col] <= percentile_75), nueva_columna] = 2
    X_test_bayes.loc[X_test_bayes[col] > percentile_75, nueva_columna] = 3

X_train_bayes = X_train_bayes.drop(columnas_numericas, axis=1)
X_test_bayes = X_test_bayes.drop(columnas_numericas, axis=1)
```
